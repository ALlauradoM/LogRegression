```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn import preprocessing
```

```python
import pandas as pd 
df = pd.read_csv('social_network_ads.csv')
#print(df.shape[0])
print(df.describe())
data = np.hstack((df['Age'].values.reshape((df.shape[0],1)), df['EstimatedSalary'].values.reshape((df.shape[0],1))))
#print (data)
X = data
y = df['Purchased'].values.reshape((df.shape[0],))
```

```python
class LogisticRegression:
    def __init__(self, lr=0.01, num_iter=100000, fit_ones=True):
        self.lr = lr
        self.num_iter = num_iter
        self.fit_ones = fit_ones
    
    def add_ones(self, X):
        ones = np.ones((X.shape[0], 1))
        return np.concatenate((ones, X), axis=1)
    
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        X = preprocessing.normalize(X)
        if self.fit_ones:
            X = self.add_ones(X)
        
        # weights initialization
        self.theta = np.zeros(X.shape[1])
        self.error = np.zeros(self.num_iter)
        
        for i in range(self.num_iter):
            z = np.dot(X, self.theta)
            h = self.sigmoid(z)
            gradient = np.dot(X.T, (h - y)) / y.size
            # Calculate Loss:
            self.error[i] = (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
            self.theta -= (self.lr * gradient)
    
    def predict_prob(self, X):
        if self.fit_ones:
            X = self.add_ones(X)
    
        return self.sigmoid(np.dot(X, self.theta))
    
    def predict(self, X, threshold):
        X = preprocessing.normalize(X)
        return self.predict_prob(X) >= threshold
```

```python
model = LogisticRegression(lr=0.1, num_iter=1000)
model.fit(X, y)
```

```python
def plot_data(X,y):
    a1 = [0, 0]
    a2 = [0, 0]
    for i in range(X.shape[0]):
        if y[i] == 0:
            a1 = np.vstack((a1, X[i,:]))
        else:
            a2 = np.vstack((a2, X[i,:]))

    plt.title("Data")
    plt.scatter(a1[1:,0], a1[1:,1])
    plt.scatter(a2[1:,0], a2[1:,1])
    plt.show()

def plot_error(error):
    plt.title("Error - Iteracions")
    plt.plot(error)
    plt.show()
    
def calculate_dec_bound(x,theta):
    y0 = ((-theta[0])-(x[0]*theta[1]))/theta[2]
    y1 = (-(theta[0])-(x[1]*theta[1]))/theta[2]
    return [y0, y1]
    
def plot_result(X,y,theta):
    a1 = [0, 0]
    a2 = [0, 0]
    for i in range(X.shape[0]):
        if y[i] == 0:
            a1 = np.vstack((a1, X[i,:]))
        else:
            a2 = np.vstack((a2, X[i,:]))

    # Decision Boundary: sum(X*theta) == 0
    x0 = [min(X[:,0]), max(X[:,0])]
    x1 = calculate_dec_bound(x0,theta)
    x2 = [(abs(x1[0])*(max(X[:,1])-min(X[:,1])))+min(X[:,1]),(abs(x1[1])*(max(X[:,1])-min(X[:,1])))+min(X[:,1])]
    
    print("x0: "+x0.__str__())
    print("x1: "+x1.__str__())
    #print("x2: "+x2.__str__())
    
    plt.title("Result")
    plt.plot(x0,x2, c='grey')
    plt.scatter(a1[1:,0], a1[1:,1])
    plt.scatter(a2[1:,0], a2[1:,1])
    plt.show()
```

```python
plot_data(X,y)

print("Theta: "+model.theta.__str__())

plot_error(model.error)

pred = model.predict(X, .5)
print ("Accuracy: "+((pred == y).mean()).__str__())

plot_result(X,y,model.theta)
```
