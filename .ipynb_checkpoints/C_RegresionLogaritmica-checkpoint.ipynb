```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import *
from numpy import where, ones, zeros, log 
from scipy.optimize import fmin_bfgs
import math
```

```python
data = pd.read_csv("social_network_ads.csv")
num_usuarios = str(len(data.index))
print(num_usuarios)
```

```python
def sigmoide(X):
    return 1/(1+np.exp(-X))
```

```python
def funcionCoste(theta,X,y):
    m = y.size
    h = sigmoide(np.dot(X,theta))
    J =(-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
    return J
```

```python
def funcionGrad(alpha,theta,X,y):
    m = y.size
    grad = zeros(3)
    h = sigmoide(np.dot(X,theta))
    grad = alpha*np.dot(X.T, (h - y)) / y.size
    return grad
```

```python
X = data.values[:, 2:4]
y = np.array(data.Purchased)

pos = where(y == 1)
neg = where(y == 0)
```
